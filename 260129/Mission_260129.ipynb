- 가격 크롤링
> .instock availability > #In stock"

```python
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/"
res = requests.get(url)
res.encoding = "utf-8"
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select(".product_pod")
base_url = "https://books.toscrape.com/"

for index, item in enumerate(items) :
    title= item.select_one("h3")
    price= item.select_one(".product_price > .price_color")
    link= item.select_one("[href]")
    total_url = base_url + link["href"]
    
    # stock= item.select_one(".product_price > .instock availability")


    print(index,title.get_text(" ",strip=True), price.get_text(" ", strip=True), total_url)
```

```python
#포스트맨을 통한 네이버 API는 최대 100개까지밖에 값을 안 줌. 우리는 1000개의 값을 원함
#=> 100개씩 찾아오는 걸 10번 반복해서 1000개 찾아올거임
# 위 데이터를 엑셀파일로 저장

import requests
import openpyxl

client_id = "iQYuA4jIAZWE78ZCH4p9"
client_secret = "7MP3lY3dzz"

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active   #active : 엑셀파일에서 기본적으로 활성화된 sheet
excel_sheet.column_dimensions["B"].width = 100
excel_sheet.column_dimensions["C"].width = 100
excel_sheet.append(["No", "제목", "링크"]) #1행  #행은 반드시 리스트값이어야함



#A열: 1~1000 , B열  : 제목, C열: 상품링크
#행 :1행내용 -> 넘버, 제목, 상품링크

#포스트맨을 통한 네이버 API는 최대 100개까지밖에 값을 안 줌. 우리는 1000개의 값을 원함
#네이버: 기본 10개, 최대 100개 => 100개씩 찾아오는 걸 10번 반복해서 1000개 찾아올거임 => 1~100 // 101~200/ 201~300 ... 901~1000
# start = 1 ~ 100 

#start 값을 1, 101,...,901 같이 어떻게 자동으로 반복할 수 있는 알고리즘을 만들거냐
start = 1
num = 0
for index in range(10) :  #index는 0부터 시작 ->0~9까지 실행됨
    start_number = start + (index * 100)

    naver_open_api = f"https://openapi.naver.com/v1/search/shop.json?query=독도토너&display=100&start={str(start_number)}"
    #display=100 : 한 번에 표시할 검색 결과 개수최댓값: 100)
    #start={str(start_number)

    header_params = {
        "X-Naver-Client-Id" : client_id,
        "X-Naver-Client-Secret" : client_secret
    }
    
    res = requests.get(naver_open_api, headers= header_params)
    
    
    if res.status_code == 200 :
       data = res.json()
       for item in data["items"] :
           num += 1   # 변수 num의 정의가 for문 밖에 있으므로 1~1000까지 감 
           excel_sheet.append(num, item["title"], item["link"])  #god
           print([num, item["title"], item["link"]])
    else :
        print("Error Code:", res.status_code)

#만들어놓은 엑셀파일에 저장
excel_file.save("shopping_with_독도토너.xlsx")
excel_file.close()
print("shopping_with_독도토너.xlsx 파일 저장 완료!")




        
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

# print(soup)

# section = soup.select(".course.maintitle")
# print(section)

section = soup.find("ul", id="dev_course_list")
titles = section.find_all("li", "course")

# enumerate

for index, title in enumerate(titles) :
    print(index + 1, "댓글 수 :", title.get_text(" ", strip=True).split(" - ")[-1].replace("\n", " ").replace(" ", "")[-2:-1])
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("ul > li")
for item in items :
    print(item.get_text(" ", strip=True).replace("\n", "").replace(" ", ""))
    
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("#dev_course_list > .course")
for item in items :
    print(item.get_text(" ", strip=True))
```

# 오늘(2026. 01. 22)의 미션!!!
해당 사이트에 접속해서 데이터를 크롤링하세요 -> https://quotes.toscrape.com/
해당 사이트에서 명언, 사람, 태그 를 크롤링하세요
위 3개 요소를 크롤링한 이후 해당 값을 딕셔너리 자료형태로 묶어주세요
해당 딕셔너리를 감싸는 리스트를 만들어서 최종적으로 출력해주세요.

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

res = requests.get(url, headers=headers)
res.encoding = "utf-8"
soup = BeautifulSoup(res.text, "html.parser")

items = []

for article in soup.select(".quote") :
        items.append({
            "quote":article.select_one(".text").text.strip(),
            "author": article.select_one(".author").text.strip(),
            "tag": article.select_one(".tag").text.strip()
        })
print(items)
```

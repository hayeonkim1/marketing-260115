## 공공데이터 사이트 API를 통한 정보 수집

```python
import requests

url = "https://apis.data.go.kr/B551011/GoCamping/basedList"
requests_params = {
    "numOfRows": "10",
    "pageNo": "1",
    "MobileOS": "ETC",
    "MobileApp": "camping",
    "serviceKey": "11f7d736a05b673007597162c81c5c1ffdecf6818e135a50188e95c9e435b1df",
    "_type": "xml"
}



response = requests.get(url, params= requests_params)   #-> params : headers=headers 떼처럼 params는 정해진 속성값
print(response)
```

```python
#위에가 안될 경우
import subprocess
import xml.etree.ElementTree as ET   #문자형태의 값을 XML형식의 태그형태로 인식할 수 있게하는 모듈 (마치 BeautifulSoup)

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=11f7d736a05b673007597162c81c5c1ffdecf6818e135a50188e95c9e435b1df&_type=xml"

result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
root = ET.fromstring(result.stdout)

print(root.findall(".//item"))
```

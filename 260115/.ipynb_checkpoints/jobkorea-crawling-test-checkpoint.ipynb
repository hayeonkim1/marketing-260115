```python
from selenium import webdriver #--> webdriver가 실질적 크롤러 역할을 함
from selenium.webdriver.chrome.service import Service #웹 브라우저 드라이브 이식하는 역할
from selenium.webdriver.chrome.options import Options #웹의 옵션들을 지정
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from urllib import parse 
import time
import pandas as pd #--> as 뒤의 별칭으로 부르겠다

service = Service(ChromeDriverManager().install()) #--> 대문자이고 복잡한 클래스 대신 변수로 지정해서 사용
options = Options()
# options.add_argument("--headless") #--> headless: 가상의 브라우저를 우리눈에 보이지 않도록 설정 (띄우지 않고 몰래 갔다오는 것)
# options.add_argument("--disable-gpu") #--> 어짜피 눈에 안보이니까 굳이 GPU사용하지 않아도 되게끔 설정
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=190*1080")
options.add_argument("--start-maximized") # 브라우저 사이즈가 작더라도 실제 pc로 인식하게끔 설정
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")

driver = webdriver.Chrome(service=service, options=options)

#어느 사이트에서 어떤 값으로 찾아올 것인가
#----------------밑에 셀에 따로 정리-----
query = "퍼포먼스마케터" 
encoded_query = parse.quote(query)


#--> 사이트 주소의 페이지 번호에 따라 맨뒤 숫자가 달라지므로 숫자를 변수화해야함.
max_pages = 6

results = []  #-> 전체 공고제목 &지원링크를 한 list 안에 넣기 /// -> 리스트 값은 나중에 append()로 뒤에서부터 넣을 거임.
for page in range(1, max_pages + 1) : 
    url = f"https://www.jobkorea.co.kr/Search?stext={encoded_query}&Page_No={page}"
    driver.get(url)
    time.sleep(2)  #-> 하나의 페이지를 갔다가 다음 페이지로 가기전에 2초를 쉬라는 것.
#-----------------메모-------------------
# f-string을 안쓸경우 url => "https://www.jobkorea.co.kr/Search?stext="+ encoded_query + "&Page_No="{page} 
#--> f string 을 쓰지 않으면 문자와 변수를 다 쪼개서 연결연산자로 연결해서 써야함-> 귀찮음
# fstring 기법을 써서 문자가 아닌 변수에만 {}를 표현하는 방식으로 하나로 합쳐서 쓰기위한 것
#----------------------------------------

    
    job_posts = driver.find_elements(By.CSS_SELECTOR, "div.styles_mb_space2__dk46ts4t") 
   # styles_mb_space2__dk46ts4t  --> 모든 모집 공고와 url을 감싸고 있는 공통적인 상위 class를 찾아야함.
    for post in job_posts :
        try : # 정상적으로 실행될 경우 아래를 진행
            title_elem = post.find_element(By.CSS_SELECTOR, "a.Flex_display_flex__i0l0hl2.Flex_gap_space6__i0l0hl1g.Flex_align_center__i0l0hl8.styles_mb_space2__dk46ts4t" )
                # -> find_element 하나만 찾아와라 
                # "a.Flex_display_flex__i0l0hl2.Flex_gap_space6__i0l0hl1g.Flex_align_center__i0l0hl8.styles_mb_space2__dk46ts4t" 찾는 법
                  #=> 검사해서 공고title에 해당하는 <span 이라고 되어있는 것을 감싸고있는 <a 부분의 class 값을 복사
            title = title_elem.text.strip()  #--> title_elem를 print하되, 볼수있게 코드가 아닌 text로, 여백없이 해라
            link = title_elem.get_attribute("href") #--> 공고 url을 가져옴
            results.append({
                "공고제목": title,
                "지원링크": link
            })
            
        except : #실행 안될 경우 아래 continue를 실행
            continue


    
driver.quit()
df = pd.DataFrame(results)  #--> 크롤링한 내용을 행과열의 형태로 표시
df.to_excel("잡코리아_퍼포먼스마케터.xlsx", index=False)  #--> 엑셀에 저장하되 인덱스 숫자는 빼기
print("잡코리아_퍼포먼스마케터.xlsx 저장완료")
```

# 파이썬 기초문법
- 출력 : print() -> 내장함수 = built in function/내장: 애초에 파이썬 문법 탄생될때부터 존재
- 입력 : input("당신의 이름은?") -> 사용자로부터 어떤 값을 입력받을 수 있음 / 인자값 =placeholder : 입력해야할 값의 정보
  - !!!! 해당 함수를 통해 수집한 값은 무조건 "문자열"이 된다 !!!!!!
- 연산 :
   > 사칙연산 : + @ -> + - * ? % ** -> '+' : 문자간의 연결연산자로서도 사용됨.!
   > 비교연산 : <, >, <=, >=, ==(같다), !=(같지 않다) /// = 한개일때:  (좌항=우항, 우항의 값을 좌항으로 삽입,적용한다) 라는 뜻
- 변수 :
   - variable => 변할 수 있는 값을 넣는 공간
   - 변수명 작명 => 직관적,명시적 -> 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 맨먼저 올 수 없음(중간에 오는 건 가능)
   - 팁 => 스네이크 표기법(뱀-> case_study) / 카멜 표기법 (낙타-> caseStudy)

- 자료형 : 변수안에 담을 수 있는 값의 형태 -> 문자열(String) " " VS 문자(Character) ' ' / 정수(Int) / 실수(Float) / 논리형(Boolean)
   - 형변환 : 자동형 변환: input() 등의 함수를 거치며 자동적으로 변환됨 / 명시적형 변환 : 수동적으로 형태를 변환
   - 자료형 확인 함수: type()
  - 명시적형 변환 종류
   - int() : 정수형 변환
   - float() : 실수형 변환
   - str() : 문자열 변환
   - bool() : 논리형 변환
   - list() : 리스트 자료구조 변환
   - dict() : 딕셔너리 자료구조 변환
   - tuple(): 튜플 자료구조 변환
   - set() : 세트(집합)자료구조 변환

- 문자열 : 독립적인 함수 혹은 속성들이 많이 존재
  - count() : 특정 문자열에서 어떤 단어가 몇번 사용되었는지 집계하는 함수 (대,소문자 구분함)
  - len() : length = 길이
  - find() : 해당 문자열에서 인자값으로 전달된 문자가 몇번째 인덱스에 위치하는지 찾고자 할 때 사용
  - replace() : 특정 문자열을 대체하고자 할 때
  - index : 문자열 안에 각각의 문자는 고유한 숫자를 가지고 있다.
    -> 문자열 왼쪽부터 숫자 시작 : 0부터 시작
  - slicing : 문자열 안에서 특정 구간을 잘라 가져오고 싶을 때.
        -> 문법: 찾아오고자하는 문자열 [start : finish +1]  #--> 인덱스가 0부터 시작되므로 원하는 번째 문자 인덱스의 +1 해야되는 것.
  - strip() : 특정 문자열 앞,뒤쪽에 여백 혹은 여백으로 인지될 수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할 때
  - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때, 사용할 수 있는 기능

- 형식 지정자 :
      - 특정 타입의 값을 출력하고자 할 때, 어떤 형태로 값을 출력할지 결정
      - %s -> string
      - %c -> character
      - %d -> int
      - %f -> float

```python
text = "banana"  #text 에 바나나 할당해서 변수 지정

index = text.find("a") 

next_index = text.find("a", index + 1)
print(index, next_index)
```

```python
#반복문 사용 while

text = "banana"
target = "a"

idx = text.find(target)  #존재하지 않는 문자를 찾아오라고 하면 -> -1 도출
#--> 예외조항 처리: 어떤 명령문을 실행시킬 때, 만약 ~가 아니라면 ,~해라!!

while idx != -1 :
    print(idx)
    idx = text.find(target, idx +1)
```

```python
name = "I have a {}, I have an {}."
print(name)

print(name.format("pen", "apple")) #--> ()안의 값은 순차적으로 {}에 들어옴
print( "I have a {}, I have an {}.".format("pen", "apple")) #--> ()안의 값은 순차적으로 {}에 들어옴
print( "I have an {1}, I have a {0}.".format("pen", "apple"))  #--> 직접 () 값의 인덱스를 {}에 넣어주면 해당 인덱스의 값이 {}에 들어감
#--> 객체- 메서드(method) 함수 #반드시 함수 앞에 "문자열 형식의 객체.format()"의 형식으로 쓰임
```

```python
interest = 0.087
print(format(interest, ".2f")) #--> 전혀 다른 독립함수임 -> 소수점 위치 일치시키는데 사용
```

```python
# 파이썬 내장 함수 중 무조건 소수점 자리를 올리거나 내리는 독립기능 없음.
# 외부 라이브러리 모듈을 찾아와서 사용
# ex) 수학 객체 라이브러리 모듈
import math
interest =0.891
result01 = math.floor(interest) #--> floor: 바닥 = 무조건 버림
print(result01)

result02 = math.ceil(interest) #ceil:천장 =무조건 올림
print(result02)
```

```python
#최종본
import os
import requests
import pandas as pd

API_KEY = os.getenv("TMDB_API_KEY", "cc75647e6193da5a2ff201cb62c3caef")
NOW_LANG = "ko-KR"
# PAGE = 1
GENRE_LANG = "ko"
POSTER_BASE = "https://image.tmdb.org/t/p/w500"
BACKDROP_BASE = "https://image.tmdb.org/t/p/w780"

# genres 데이터 적용하기
def genre_ids_to_names(ids, genre_map: dict) -> str :
    # [28, 53, 80] -> '액션|스릴러|범죄'

    if not isinstance(ids, list) :
        return ""

    names = [genre_map.get(i, str(i)) for i in ids] # ["액션", "스릴러", "범죄"]
    return "|".join(names) # '액션|스릴러|범죄'

# 텍스트 개행 수정 함수
def clean_one_line(text: str) -> str :
    if not text :
        return ""
    text = text.replace("\n", " ")
    return " ".join(text.split())

# genres 데이터 가져오기
def fetch_genre_map(api_key: str, language: str = "ko") -> dict :
    url = "https://api.themoviedb.org/3/genre/movie/list"
    params = {"api_key": api_key, "language": language}
    res = requests.get(url, params=params, timeout=20)
    res.raise_for_status()

    genres = res.json().get("genres", [])
    return {g["id"] : g["name"] for g in genres} # {28 : "액션", 12: "모험", ...}

# genres 값 찾아오기
genre_map = fetch_genre_map(API_KEY, GENRE_LANG)

# now playing 데이터 가져오기
url = "https://api.themoviedb.org/3/movie/now_playing"
rows = []

for page in range(1, 6) :

    params = {"api_key": API_KEY, "language": NOW_LANG, "page": page}
    
    res = requests.get(url, params=params, timeout=20)
    res.raise_for_status()
    
    data = res.json()

    for m in data.get("results", []) :
        poster_path = m.get("poster_path")
        backdrop_path = m.get("backdrop_path")
    
        poster_url = f"{POSTER_BASE}{poster_path}" if poster_path else ""
        backdrop_url = f"{BACKDROP_BASE}{backdrop_path}" if backdrop_path else ""
    
        genre_ids = m.get("genre_ids", [])
    
        rows.append({
            "id": m.get("id"),
            "title": m.get("title"),
            "original_title": m.get("original_title"),
            "overview": clean_one_line(m.get("overview")),
            "genre_ids": ",".join(map(str, genre_ids)) if isinstance(genre_ids, list) else "",
            "genre_names": genre_ids_to_names(genre_ids, genre_map),
            "adult": m.get("adult"),
            "original_language": m.get("original_language"),
            "vote_average": m.get("vote_average"),
            "popularity": m.get("popularity"),
            "vote_count": m.get("vote_count"),
            "poster_url": poster_url,
            "backdrop_url": backdrop_url
        })

df = pd.DataFrame(rows, columns=[
    "id",
    "title",
    "original_title",
    "overview",
    "genre_names",
    "genre_ids",
    "adult",
    "original_language",
    "vote_average",
    "popularity",
    "vote_count",
    "poster_url",
    "backdrop_url"
])

out_path = "tmdb_now_playing.csv"
df.to_csv(out_path, index=False, encoding="utf-8-sig")
print(f"✅ 저장완료 : {out_path} (rows={len(df)})")

df.head()

# json, xml, csv
# data.get("dates")
# """
# id title original_title adult -> field명 그룹 => columns
# 1   더립    the rip       false

# """
```

```python
import pandas as pd

CSV_PATH = "tmdb_now_playing.csv"

df = pd.read_csv(CSV_PATH, encoding="utf-8-sig")

# 시각화 => 차트/도표로 만든다는 것
# 데이터 => 수치화 필요 (숫자)
#EX) 남,여 => 남=0, 여=1

#표의 숫자가 문자일 수 있기때문에 정수로 변환하기
num_cols = ["vote_average","popularity","vote_count"]  #-> 표의 필드명의 모음은 [, , ,]로 감싸서 표시

for c in num_cols :
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")   #-> to_numeric():  ()안의 값을 수치형 데이터로 바꿔주는 함수


#장르를 | 단위로 끊어서 3개의 장르로 된 하나의 영화를 각 장르별로 3번 중복해서 나타냄.
def split_genres(x: str) :
    x = "" if pd.isna(x) else str(x).strip() # is-> true/false 판단 함수기능 , #na = not available= 사용할 수 없는 값=결측치
    if not x :  #-. 쓸 수 없는 값이 아니라 값이 아예 없다면 이라는 뜻
        return []
    if "|" in x:
        parts = [p.strip() for p in x.split("|")]
    else :
        parts = [p.strip() for p in x.split(",")]
    return [p for p in parts if p] 

df_g = df.copy()
df_g["genre"] = df_g["genre_names"].apply(split_genres)  #.apply()  : ()안의 함수를 . 앞에 인자값으로 적용하겠다
df_g = df_g.explode("genre")
df_g

#데이터를 장르로  묶어버리기
agg = (
    df_g.groupby("genre", as_index=False)
        .agg(
            movie_count=("id","nunique"),
            avg_popularity=("popularity","mean"),
            avg_vote_average=("vote_average","mean"),
            avg_vote_count=("vote_count","mean")
        ) #grouping은 하지만 index로 쓰지는 않기 위해 index 지정
) 

agg
#aggregation : 집계
```

```python
#데이터 시각화-plotly사용

import plotly.express as px

fig = px.bar(
    agg.sort_values("movie_count", ascending=False),  #-> movie_count를 기준으로 정렬해줘(디폴트: 오름차순(=ascending))
    x="genre",  #x축,y축
    y="movie_count",
    title="장르별 영화수 분포(기준: Now Playing)"
)

#각 축의 이름 재설정
fig.update_layout(xaxis_title="장르",yaxis_title="영화 수")

fig.show() #출력
```

## Boxplot 이해하기
- 중앙값이 밑에있다: 상대적으로 큰값이 위에 있다

- 하위 50% : median을 기준으로 밑의 값
- 상위 50% : median을 기준으로 위의 값

-  Q1 = 하위 50%의 중앙값 = 1사분위 = 하위 25%
-  Q3 = 상위 50%의 중앙값 = 3사분위 = 상위 25%

-IQR(Inter Quartile Range) =중앙값 기준 퍼짐정도 = Q3 - Q1 

-Lower Fence = 하한치 = Q1 - IQR x 1.5
-Upper Fence = 상한치 = Q3 + IQR x 1.5
 => 이 fence를 넘어가면 이상치라고 판단

***-min = 하한치를 벗어나지 않은 값 내에서의 최소값 
***-max = 상한치를 벗어나지 않은 값 내에서의 최대값

-mean : 평균
-편차 = 평균에서의 차이
    -> 평균으로부터 얼마나 떨어져있는가를 알고싶은 것. +- 여부가 중요한게 아님.
-표준편차 = 모든 편차들의 평균 -> but 음수가 있으면 평균 무의미
-제곱편차 = 편차값의 제곱값  => 평균 구하기 가능
-모(집단)분산 = 제곱편차 / n  => 총 데이터갯수가 확정일 경우 사용
-표(본)분산 = 제곱편차 / n-1 = => 데이터 모집단수 1000만개 중 표본 개만 뽑아서 계산  => 과소추정법칙
-표준편차 : 루트 (모분산) 
-표준편차 : 루트 (표분산)

### 예시
[12, 15, 18, 20, 22, 25, 30, 35, 40, 100]
- 100을 제외한 나머지는 50미만에 몰려있음 -> 단순평균값은 무의미
- 일반적으로 데이터 분석 시, 평균//// 중앙값(:물리적인 가운데 값)=median 사용
- 짝수: 5,6번째 값의 평균
- 홀수: 실제 가운데 1개의 값
=> median = (22+25)/2 = 23.5
-하위 50%: 12,15,18,20,22
->Q1 = 18 (하위 25%)
-상위 50%: 25,30,35,40,100
-> Q3 = 35 (상위 25%)

IQR = Q3 - Q1= 35-18 = 17
Lower Fence = 18 - (17*1.5) = -7.5
Upper Fence = 35 + (17*1.5) = 60.5

min = 12
max = 40
-> 이상치 = 100 => Outlier

mean = 12+15+18+20+22+25+30+35+40+100 = 31.7
첫번째값의 편차: 12 - 31.7 = -19.7

# 히트맵차트 : 색상 사용정도 분츄 : 색상 진해지면, 색상 연해지면
# 상관관계 : 피어슨 상관계수 -> 공분산,표준편차,분산


```python
#특정 장르등의 집합( 액션,스릴러,범죄) /(액션, 판타지, 스릴러)
# 액션 장르는 주로 어떤 장르와 많이 섞일까? 알아보기
# 범죄 장르는 어떤 장르와 혼합하는가?
#액션 , 범죄 장르의 상관관계 여부
#스릴러, 드라마 상관관계 여부


#다시 id를 기준으로 장르를 묶을 것. => 그루핑
# (액션,스릴러,범죄) => 100개의 영화 가운데 몇번 등장하는가?
#BUt 장르 혼합의 기준이 다름 
#=> 모든 영화장르를 2개의 조합으로 만들 것. (a,b)
# 교차되는게 많을수록 가중치 값을 높일 것.
```

```python
#장르별 상관관계 그래프(히트맵) 만들기

import pandas as pd
import plotly.express as px
from collections import Counter #몇번 사용됐는지 알 수 있게 해주는 Class
from itertools import combinations

movie_to_genres = (
    df_g.groupby("id")["genre"]  #-> id를 기준으로 그루핑을 하되 장르값을 찾아오자
    .apply(lambda s: sorted(set([x for x in s if isinstance(x, str) and x.strip()])))  #-> s: 1개의 id당 해당 장르들이 list형태로 묶여있는 값
    .to_dict()
) 

pair_counter = Counter()
for genres in movie_to_genres.values() : 
    #모든 장르를 2개 1쌍으로 통일
    for a, b in combinations(genres, 2) : #-> genres의 리스트 값 중 2개의 조합으로 가져와서 각 a,b로 지정
        pair_counter[(a, b)] += 1
        pair_counter[(b, a)] += 1  #-> 순서가 바껴도 가중치가 값아지도록

#값을 적용할 바둑판 만들기
all_genres = sorted(set(df_g["genre"].dropna().unique())) #dropna() 결측치 버리기   #-> unique : 중복되는 값은 하나로 합치기
mat = pd.DataFrame(0, index= all_genres, columns= all_genres, dtype=int) #바둑판 표의 세로필드 :index, 가로축값: columns #-> 가로,세로 축에 모든 장르 값 입력

for (a,b), v in pair_counter.items(): 
    if a in mat.index and b in mat.columns : 
        mat.loc[a,b] = v

fig = px.imshow(
    mat,
    title="장르조합 Heatmap"
)
fig.update_layout(xaxis_title="장르", yaxis_title="장르")
fig.show()


# """
# def 함수이름() :
#     실행문
# """

# """
# 실무에 나가는 순간 -> lambda => 함수 구문x  -> 식
# """
```

```python
#국가(언어)별 인기 장르
import numpy as np
import plotly.express as px

lang_top_k = 8

tmp = df_g.copy()
tmp["original_language"]= tmp["original_language"].fillna("unknown") 

top_langs = tmp["original_language"].value_counts().head(lang_top_k).index
tmp["lang_group"] = np.where(tmp["original_language"].isin(top_langs), tmp["original_language"], "Others") #-> original 언어가 top8개 안에 들어가면 그 언어를 쓰고 그렇지 않을 경우 "Others"를 써라

pivot = tmp.groupby(["genre", "lang_group"])["id"].nunique().reset_index(name="movie_count") #장르로 1차 그룹핑, 그 안에서 lang_group으로 2차 그룹핑 

fig = px.bar(
    pivot,
    x="genre",
    y="movie_count",
    color="lang_group",
    title="장르별 언어분포"
)
fig.update_layout(xaxis_title="장르", yaxis_title="영화 수")
fig.show()
```

```python
price = "100 달러" 
exchange_rate = 1480

#price라는 변수에 있는 100달러의 값을 환율을 적용해서 한화(한국돈)으로 변환해서 값을 출력해라


# price01 = price.split(" ")[0]
# print(price01)


# 문자열: 인덱스 -> 슬라이싱
print(int(price[0:3]))

result = int(price[0:3]) * exchange_rate
print(f"{result:,}원")  #--> :, : 1000의 단위마다  , 를 찍어라 

```

```python
# 반복문을 활용하여 구구단 2~9단까지를 아래에 출력해주세요

#2~9단 반복

for index_1 in range(2,10):
    print(f"====구구단 {index_1}====")
    for index_2 in range(1,10) : 
    
        print(f"{index_1}x{index_2}= {index_1 * index_2}")
    print("\n") #하나가 끝나면 행을 바꾸고 escape 해라 


    
```

# 파이썬 함수 (function)
- 함수 :  어떤 읿력값을 받아서 내부에서 연산처리 후, 해당 결과값을 외부로 출력할 수 있도록 기능하는 것
  -> 프로그래밍 언어의 꽃
- 함수 값을 입력 & 출력: input & output
- 함수 생성(=선언)->  함수 존재 -> 값 입력(입력하는 값: 인자값 or 인수 -> 매개변수) -> 값이 없을 수도 있음.
- 함수 사용(=호출) -> print 없이 함수만 입력
- 함수의 선언부의 인자값 개수와 호출부의 인자값의 갯수는 값아야함. (어느 한쪽이 상대적으로 갯수가 많거나 적으면 정상적으로 실행X)
- return문 : 함수의 내부연산 처리 후, 특정값을 함수 바깥으로 값을 보낼 수 있는 기능. 

def 함수1() :
    실제 함수의 실행이 진행될 공간 -> 반복문, 조건문처럼 들여쓰기!!

# 파이썬 자료구조

- 논리형 = True

- 리스트 [] = [1, 2, 3]  -> 원하는 값 추가/삭제 가능
- 튜플 {} = (1, 2) ->(리스트와 거의 동일) -> 최초 선언된 형태의 자료구조 변경 불가. 데이터 추가/삭제 불가능
> iter & 반복문 사용가능
> 리스트와 튜플 안에 담겨있는 값들이 어떤 값을 의미하는지 알 수 없음.
- 딕셔너리() = 사전 = {"apple" : "사과"} -> {key : value}라고 부름 -> 한쌍의 {key: value} => property = 속성
- 집합 =수학시간 배웠던 집합 개념
> 교집합,합집합,차집합,공집합 등등
> 반복되는 값 허용하지 않음. -> 데이터 크롤링 값이 반복적으로 입력된 값을 1회만 처리할때

```python
def test_func1(a,b) :
    quot = a // b # // :몫
    rem = a % b # % 나머지
    return (quot, rem)
```

```python
result = test_func1(10, 3)


# 구조분해할당 기능 
(f, s)= result
print(result)
# (f, s) = (3, 1)
print(f, s)
```

```python
data_dict = {"한국": "KR", "일본": "JR", "중국": "CN" }
print(type(data_dict))
```

```python
#list, tuple => 내부 아이템에 고유한 인덱스 0 ~ 가 존재
# print(dir(data_dict))
# data_dict[0] -> 오류 
# 딕셔너리는 인덱스가 아닌 key,value로 사용

data_dict["한국"]  #-> []에 key값을 입력하면 value 값이 출력됨!!!
print(data_dict["한국"], data_dict["일본"])
```

```python
# 딕셔너리: [] 안의 key가 없으면 "" 안의 value 값을 가지는 한 속성을 추가할 수 있음

data_dict["미국"] = "US"  
data_dict
```

```python
#재할당도 가능
data_dict["미국"] = "USA"  
```

```python
#딕셔너리는 iter 속성을 가지고 있다. 
for item in data_dict :
    print(item, data_dict[item])
    
for item in data_dict :
    print(f"{item} {data_dict[item]}")


# key, value => 반복문이 아니면 못찾아오나? NO 가능함
```

```python
dir(data_dict) 
# dictionary의 속성 중 
#keys, values, items 요소. -> iterable 함 => 반복문 가능
```

```python
#key 만 찾아오기
data_dict.keys() 

#반복문
for keys in data_dict.keys() :
    print(keys)
```

```python
# value만 찾아오기
data_dict.values()

for values in data_dict.values() :
    print(values)
```

```python
# item(둘다) 찾아오기
data_dict.items()

for items in data_dict.items() :
    print(items)
```

```python
#딕셔너리와 리스트 목적차이

# ex) 인터파크 -> 공연 페이지의 데이터 수집 크롤링
# 리스트형태 사용 좋지 않음 ["워터밤", "사랑합니다"]-> 공연명 -> 그 가수를 알수 없음
#딕셔너리 사용 {"싸이": "워터밤", "성시경": "사랑합니다"}  => 하나의 쌍으로 가져올 수 있음.

#튜플은 거의 사용하지 않음.
```

```python
data1 = {"apple","samsung","lg"}
data2 = {"xaiomi","samsung","lg"}
```

```python
# 교집합
data1 & data2
```

```python
# 합집합
data1 | data2
```

```python
#차집합
data1 - data2
```

```python
# 교집합을 제외한 나머지 부분집합
data1 ^ data2
```

```python
# set 자료구조는 중복값을 허용하지 않는다!! 
data_list = ["apple","dell", "samsung", "lg", "lg", "apple", "dell", "apple", "samsung"]
data_list
```

```python
set_result = set(data_list)   #--> 명시적 형변환 VS 자동 형변환
print(set_result)
```

```python
# 특정 사이트에서 데이터 크롤링 > 대부분 리스트 형태 > 중복값 필연적 > 세트 자료구조로 변환 후 > 다시 리스트
```

```python
#set을 다시 list로 
list(set_result)
```

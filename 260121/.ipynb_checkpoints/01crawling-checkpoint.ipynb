```python
import requests #데이터를 수집해오는 크롤러
import bs4 #찾아온 데이터를 보다 쉽게 이해,해석하도록 아름답게 바꿔주는 역할

url = "https://news.daum.net/tech"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.encoding = "utf-8"  #-> 한국어-> 컴퓨터가 이해할 수 있는 언어로 변환해라
soup = bs4.BeautifulSoup(response.text, "html.parser")

articles = soup.select("ul.list_newsheadline2 li a.item_newsheadline2") #list_newsheadline2 라는 ul앵커태크 의 자식 li 의 자식 a 앵커태그를 select해라

for article in articles :
    title = article.text.strip()
    link = article["href"]
    print(f"{title} - {link}")

print(articles)
```

```python
#위키피디아 사이트 > 마케팅 검색 후 결과값을 크롤링

import requests
import bs4
url = "https://ko.wikipedia.org/wiki/%EB%A7%88%EC%BC%80%ED%8C%85"
headers = {
     "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = bs4.BeautifulSoup(response.text,"html.parser")

paragraph = soup.select("div.mw-content-ltr.mw-parser-output > p")

for p in paragraph[:2] :
    print(p.get_text().strip())
```

```python
import requests
import bs4
url = "https://www.jobkorea.co.kr/Search/?stext=%EC%BB%A8%ED%85%90%EC%B8%A0%EB%A7%88%EC%BC%80%ED%84%B0"
headers= {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = bs4.BeautifulSoup(response.text,"html.parser")
articles = soup.select("div.styles_mb_space2__dk46ts4t a ")

for article in articles[:10] :
    title = article.text.strip()
    link = article["href"]
    print(f"{title} - {link}")
            
```

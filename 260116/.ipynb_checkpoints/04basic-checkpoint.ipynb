# 데이터 자료구조 1 - 리스트 (list)
- 문자열, 정수, 실수, 논리형 -> 변수 1개당 값 1개 (1:1) -> 너무 많은 변수가 필요해짐
- 변수 1개 : 여러개의 값을 그룹화
      -> 공통분모를 가지고 있는 값들의 모임-> 특성분류
      -> 상대적으로 효율적인 컴퓨터 메모리 공간 사용
      -> 대부분의 크롤링 결과값이 리스트 형태의 자료구조로 값이 결정됨
- 형태: 변수명 = [값, 값, 값]
- 리스트 & 문자열 -> 유사함 -> 비슷한 문법 구조
      ->리스트, 문자열 -> 공통 속성을 가지고 있기 때문
      -> __iter__ (=iterable): 반복할 수 있는 / index를 카운트하는 방식이 같음

-고유 문법: 
    - 값을 추가하는 또다른 방법
        -> append() : 리스트에 값을 추가 (뒤에서부터)
        -> list + list 연산
        -> insert(추가할 인덱스 , 넣을 값) 
        -> extend(여러개의 값을 한번에 추가하고자 할때)
    - 값을 제거하는 방법
        -> pop() : 리스트의 값을 제거 (뒤에서부터) -> append()의 반대=> 인자값으로 인덱스값 부여할 경우 -> 해당 인덱스의 값이 제거
        -> remove() : 리스트에서 제거할 특정 값을 직접 선택 가능
        -> del 리스트명[인덱스] : 해당 인덱스 값 제거
    - 슬라이싱 기능 = 문자열과 동일 작동
    - 리스트 자료구조안에 입력 값의 타입이 상이해도 가능
    - 리스트 안에 담겨있는 값을 정렬 (오름차순/내림차순)
        - sort(): 오름차순 정렬 함수 (일시적 정렬이 아닌 원본 데이터를 편집하는 기능)
        - reverse(): 내림차순 정렬 함수 (원본 데이터 편집 기능)

- 문자열을 리스트로 변경해주는 기능 : split()
  왜 바꿔야 하는가? 외부정보를 가져왔을 때 문자열 형태일때가 많음 -> 그 정보를 쪼개야할때 리스트로 변경해야 쪼개기.
  => 반복문도 사용 가능해짐
  -> 원본 데이터를 편집할 수 없음
  

```python
numbers = [2, 1, 4, 3]
# print(type(numbers))
print(numbers)
```

```python
numbers.sort() #--> sort() : 오름차순으로 정렬하는 함수 
print(numbers)
```

```python
print(numbers)  #--> sort는 일시적 정렬이 아닌 원본 데이터를 편집하는 기능
```

```python
numbers.reverse()
print(numbers)
```

```python
print(numbers)
```

```python
words = "python is easy"  #-> 문자열을 변수 지정
results = words.split()
for item in results :
    print(item)
```

# 조건문: 값을 출력하기 위해서, A | B 에 따라서 다른 값을 출력해야할때

1) if 조건문 : A라면 ~해라  -> 단일조건
2) if else문 : A라면 ~하고, 만약 A가 아니라면 ~ 해라 (양자택일)
3) if elif else문 : A라면~ ,B라면~,C라면 (다중조건->여러가지 후보를 만들 수 있음)
-> else 문은 절대 if, elif보다 먼저 나올 수 없음!!
-> else 문은 (필수가 아닌)

- 조건문 : 문법 (조건식 + 실행문)
- 조건문 > 조건식 > 비교연산자 | 논리연산자
if 조건식 :
    실행문  
    ===> 무조건 조건식보다 실행문이 들여쓰여야함

if 조건1 and 조건2 :   -> 논리연산자- 1도 2도 참이라면
    실행문

if not 조건 :     --> 조건식의 값이 부정(False)인 경우에 실행문을 실행시켜야하는 경우
    실행문

if 조건식1 : ex) 돈까스 먹을래?
    실행문1          -> 기름진거 싫어
else :
    if 조건식2 : 기름지지 않은 것중에 초밥먹을래? 
        실행문2      -> 느끼해
    else :       그럼 국밥먹자
        실행문3

```python
age = input("당신의 나이는?")  #--> input()의 결과값은 무조건 문자!
age_digit = int(age) #--> digit: 10진법 #--> int(): 형변환

if age_digit >= 19 :
    print("당신은 성인입니다")
```

```python
age1 = input("흥민이의 나이는?")
age_digit1 = int(age1)


age2 = input("강인이의 나이는?")
age_digit2 = int(age2)


if age_digit1 >= 19 and age_digit2 >=19 :     #--> 논리 연산자 : 좌항과 우항 모두 참이어야함
    print("두사람 모두 성인입니다.")
```

```python
#성인이냐 아니냐 -> 양자택일이 가능 => if else문 사용
age = int(input("당신의 나이는"))

if age >= 19 :
    print("당신은 성인입니다.")
else :
    print("당신은 미성년입니다.")
```

```python
#노인/장년/청년/청소년/유소년 -> 분류를 쪼개야할 때 => if elif else 문
age = int(input("당신의 나이는"))

if age >= 19 :
    print("당신은 성인입니다.")
elif age>= 13 and age < 19 :
    print("당신은 청소년입니다.")
elif age>= 5 and age < 12 :   #-> 다중조건은 elif로 반복하다가 마지막에 else
    print("당신은 유소년입니다.")
else : #--> '그밖에' 라는 뜻 => 가장 마지막 조건문이 되야함. 
    print("당신은 유아동입니다.")
```

```python
# 사용자로부터 값을 받아서 짝수인지 홀수인지 판단 후 짝수라면 "짝수입니다." 출력, 홀수라면 "홀수입니다." 출력

number = int(input("숫자는 얼마인가요?"))

#-> 짝수: 2로 나누어떨어짐

if number % 2 == 1 :    #--> 숫자를 2로 나눈 나머지가 1이면
    print("해당 숫자는 홀수입니다.")
else :
    print("해당 숫자는 홀수입니다.")
```

```python
#더 효율적으로 1차
numbers = [] #나중에 append로 추가할거임

#반복문으로
for i in range(3) :   #--> 반복할게 없어도 쓸 수 있는 반복문
  number = int(input(f"{i + 1}번째 숫자 : "))
#-> 값이 들어올 때마다 어디든 저장을 해놔야 반복할때 덮어쓰지 않을 수 있음 => 비어있는 list 만들기
  numbers.append(number)

#구조분해할당: 리스트와 같이 여러개 변수값을 가지고 있는 자료구조의 값들을 하나씩 추출 > 독립변수로 담을 수 있는 문법
num01, num02, num03 = numbers

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
else :
    print(num03)
```

```python
#더 효율적으로
numbers = []

#반복문=반복 (문:식+식)
for i in range(3) :  
  number = int(input(f"{i + 1}번째 숫자 : "))
  numbers.append(number)

num01, num02, num03 = numbers

print(min(numbers))
```

```python
# 리스트 컴프리핸션 = list comprehension
#-> 식들의 합인 "문"을 한 문장으로 써줄게

numbers = [int(input(f"{i + 1}번재 숫자: ")) for i in range(3)]
print(min(numbers))
```

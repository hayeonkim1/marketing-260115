- 200 : 클라이언트의 요청에 서버가 문제없이 통신했다라는 뜻.
- 300 : 클라이언트의 요청에 서버가 통신을 했으나 다른곳으로 우회했다는 뜻.
- 400 : 클라이언트의 요청이 잘못된 경우, 서버에서 정상적으로 응답할 수 없다는 뜻. -> 404
- 500 : 클라이언트의 요청은 문제가 없으나, 서버에서 응답을 정상적으로 하지 못하는 경우. ex) 티켓팅 때 서버가 터지는 경우

```python
# 1) 


import requests
from bs4 import BeautifulSoup 

# 크롤링을 위해 변수에 사이트 주소를 문자형태로  url 에 할당하기.
url = "https://www.yes24.com/product/category/steadyseller?pageNumber=1&pageSize=24&categoryNumber=001001025007"

# 유저 신분증 제시
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
} 
# 딕셔너리 =사전 =단어명칭 : 뜻
# f12 누르고 페이지 새로고침 후 Network의 가장 상단의 'search.naver?where~~' 클릭 후, Header 의 "User-Agent"의 정보를 복붙 


# ---> requests한 import에 대한 response => get(가져와라. url(브라우저)와 headers(유저)의 통신상태를)
response = requests.get(url, headers)  
response.encoding = "utf-8"  # 유니코드 미국-> 영어사용 // 웹사용 국가를 파악 -> 해당 국가의 타언어를 인식할 수 있게끔 코드를 지정하는 것
                            #-> 주로 이모지,텍스트활용 이모티콘등이 깨지지 않을 수 있도록 인식시켜주는 것. 


soup = BeautifulSoup(response.text, "html.parser")   #--> 읽기 아름다운 형식으로 다듬기
print(soup) 


# 2)
# 마우스 우클릭해서 title, author에 해당하는 tag 선택하기

#---------------반복하기 싫어서 title이랑 author을 모두 포함하는 태그 "item_into"를 찾아서 사용 ---------- 
books = soup.select("div.item_info")
# for book in books[5:11] : # --> [? :? ] : 슬라이싱 -> ?번째부터 ?번째를 찾아와라
#     title_tag = book.select_one("a.gd_name")   # --> select()는 모든 리스트를 찾아옴
#     author_tag = book.select_one("span.authPub info_auth") #---> select_one() 위에서부터 최초의 값 하나를 찾아옴
#     print(title_tag.text.strip(), author_tag.text.strip())

for i, book in enumerate(books[5:11], 1)  #-->번호 지정
    title_tag = book.select_one("a.gd_name")   
    author_tag = book.select_one("span.authPub.info_auth") 
    print(f"{i}. {title_tag.text.strip()} - {author_tag.text.strip()}")
# ---------------------------


# title_tag = soup.select("a.gd_name")
# author_tag = soup.select("span.authPub info_auth")

# print(title_tag, author_tag)
# print(type(title_tag))

#print(title_tag.text) --> text만 보고싶었는데 오류 => # 문자열에서만 사용 가능
#print(type(title_tag)) #but, title_tag가 문자열(str)이 아니기 때문에 불가능한 것

# title_tag의 정체 파악하기 -> dir : 해당 변수가 쓸 수 있는 함수/속성을 싹 정리해서 보여줌.
# print(dir(title_tag))  #--> iter 사용할거임

# for title in title_tag :  # --> title_tag 속 title만 찾아보고  
#     print(title.text)  #-->그 후에 text만 가져오자. 

# for author in author_tag :
#     print(author.text.strip()) 
#     #엔터를 치는 순간 => |n : escape sequesnce : n-> 일반적인 스펠링문자, 개행처리=행바꿈
#     #strip(): 행바꿈 없이 문자만 가져오기

```

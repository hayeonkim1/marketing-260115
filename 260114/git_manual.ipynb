# git 사용방법
1) 로컬 컴퓨터 : git 응용프로그램 설치
2) 터미널 : git-v 명령어
- 검색바 > cmd 입력 > 명령 프롬포트 클릭하면 뜨는 검은 창 =터미널
- git -v 입력했을 때 git version이 나오면 있는것

3) 없는 경우, 구글크롬 >Standalone Installer -> Git for Windows/x64 Setup 눌러서 git 다운로드 및 설치
-> 터미널 껐다 키기
4) 깃의 소유주 확인 & 변경
- git config --list
      -user.email:
      -user.name:
-변경방법: 터미널 재접속 > 검색바 cmd > 관리자 권한으로 실행 : 전역(global) 세팅
-관리자 터미널에 입력 
    - git config --global user.name "hayeonkim1"
    - git config --global user.email "a2430498@gmail.com"
- git config --list 로 확인


5) github 회원가입 : 그 계정정보 값으로 입력
  -git: 나의 로컬 컴퓨터에서 버전 관리하는 툴
  -github: 내 로컬 컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할 때
           또는 내가 사용하고 있는 디바이스가 2대 이상인 경우, A에서도 데이터 관리,B에서도 데이터관리 가능
- repository : github > project에 따른 개별 폴더

6) git commit & push
- cd "현재 나의 루트폴더"이동
- git 초기화 : git init
